## 2.4. Handling missing data

IMPORTANT: We decided to limit our test and training data to those where 'Sunshine' variable is not null in order to reduce the amount of data used and to lower the number of null values in whole dataset.


```python
rain_df = rain_df.dropna(subset=['Sunshine'])
```

```python
# Check for missing data

rain_df.isna().sum()
```

```python
# fill missing data for numeric attributes with mean and median (if there are outliers [max values are much bigger 
# than mean values] in our attribute we will use median)

rain_df['Temp3pm'].fillna(rain_df['Temp3pm'].mean(), inplace=True)
rain_df['Temp9am'].fillna(rain_df['Temp9am'].mean(), inplace=True)
rain_df['Cloud3pm'].fillna(rain_df['Cloud3pm'].mean(), inplace=True)
rain_df['Cloud9am'].fillna(rain_df['Cloud9am'].mean(), inplace=True)
rain_df['Pressure3pm'].fillna(rain_df['Pressure3pm'].mean(), inplace=True)
rain_df['Pressure9am'].fillna(rain_df['Pressure9am'].mean(), inplace=True)
rain_df['Humidity3pm'].fillna(rain_df['Humidity3pm'].mean(), inplace=True)
rain_df['Humidity9am'].fillna(rain_df['Humidity9am'].mean(), inplace=True)
rain_df['WindSpeed3pm'].fillna(rain_df['WindSpeed3pm'].median(), inplace=True)
rain_df['WindSpeed9am'].fillna(rain_df['WindSpeed9am'].median(), inplace=True)
rain_df['WindGustSpeed'].fillna(rain_df['WindGustSpeed'].median(), inplace=True)
rain_df['Sunshine'].fillna(rain_df['Sunshine'].mean(), inplace=True)
rain_df['Evaporation'].fillna(rain_df['Evaporation'].median(), inplace=True)
rain_df['Rainfall'].fillna(rain_df['Rainfall'].median(), inplace=True)
rain_df['MaxTemp'].fillna(rain_df['MaxTemp'].mean(), inplace=True)
rain_df['MinTemp'].fillna(rain_df['MinTemp'].mean(), inplace=True)
```

## 2.6. Data Simplification

```python
# Since the wind direction were wide spread and very specific we decided to simplify the data by puting them into 
# categories based on dominant direction, filled the missing data for wind direction using mode.

rain_df['WindDir9am'].replace({'NNW':'N', 'NNE':'N', 'NE':'N', 'ENE':'E', 'ESE':'E', 'SE':'E', 'SSE':'S', 'SSW':'S', 'SW':'S', 'WSW':'W', 'WNW':'W', 'NW':'W'},inplace=True)
rain_df['WindDir3pm'].replace({'NNW':'N', 'NNE':'N', 'NE':'N', 'ENE':'E', 'ESE':'E', 'SE':'E', 'SSE':'S', 'SSW':'S', 'SW':'S', 'WSW':'W', 'WNW':'W', 'NW':'W'},inplace=True)
rain_df['WindGustDir'].replace({'NNW':'N', 'NNE':'N', 'NE':'N', 'ENE':'E', 'ESE':'E', 'SE':'E', 'SSE':'S', 'SSW':'S', 'SW':'S', 'WSW':'W', 'WNW':'W', 'NW':'W'},inplace=True)

rain_df['WindDir9am'].fillna(rain_df['WindDir9am'].mode()[0], inplace=True)
rain_df['WindDir3pm'].fillna(rain_df['WindDir3pm'].mode()[0], inplace=True)
rain_df['WindGustDir'].fillna(rain_df['WindGustDir'].mode()[0], inplace=True)

rain_df.tail()
```

# 5. Split train and test

# 6. Logistic Regression

```python
#first we'll check null accuracy (accuracy when predicting the most frequent value of target variable) 
#of whole dataset before balancing

rain_df['RainTomorrow'].value_counts()
```

```python
#print('Null accuracy:' ,'{:.4f}'.format(110316/(110316+31877)))
print('Null accuracy:' ,'{:.4f}'.format(57556/(57556+16821)))
```

## 6.1. Simple model training and evaluation

```python
#Simple logistic regression 
classifier =  LogisticRegression(max_iter=1000, random_state=111)
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
```

```python
log_accuracy=accuracy_score(y_test, y_pred)
log_recall=recall_score(y_test,y_pred)
log_precision=precision_score(y_test,y_pred)
log_rocauc=roc_auc_score(y_test,y_pred)
```

```python
print('{:.4f}'.format(log_accuracy), '- Log Accuracy')
print('{:.4f}'.format(log_recall), '- Log Recall')
print('{:.4f}'.format(log_precision), '- Log Precision')
print('{:.4f}'.format(log_rocauc), '- Log ROC AUC')
```

```python
print('Training set score: {:.4f}'.format(classifier.score(x_train, y_train)))

print('Test set score: {:.4f}'.format(classifier.score(x_test, y_test)))

print('Null accuracy:' ,'{:.4f}'.format(57556/(57556+16821)))
```

Our model turns out to be only slightly better in accuracy than choosing most frequent value. Also we could be dealing with underfitting (train and test scores are really close).

## 6.2. Model training with hyperparameter tuning

```python
log_C = [0.01 , 0.1 , 0.5 , 1 , 10 , 25 , 50 , 100]
log_solvers = ['liblinear' , 'newton-cholesky']


log_grid = {'C': log_C,
               'solver': log_solvers
            }
print(log_grid)
```

```python
#Grid search with CV
log_gridsearch = GridSearchCV(estimator = classifier, param_grid = log_grid, 
                                cv = 8, verbose = 2,
                                n_jobs = -1, refit=True)
```

```python
log_gridsearch.fit(x_train, y_train)
```

## General conclusion

### Exploration
After uploading the dataset we checked the target variable **Rain Tomorrow** we noticed a class imbalance in it and we deleted the rows with missing target variable, further checked the data types for variables in the original dataset in which we found that there is: **1 Date:** Date. **5 Categorical:** Location, WindGustDir, WindDir9am, WindDir3pm, RainToday. **17 Numerical:** MinTemp, MaxTemp, Rainfall, Evaporation, Sunshine, WindGustSpeed, WindSpeed9am, WindSpeed3pm, Humidity9am, Humidity3pm, Cloud9am, Cloud3pm, RainTomorrow.

### Preprocessing
We checked for missing data using *isna().sum()* and invalid data by analysing the results from *decribe()* we decided to drop the missing data rows for **Sunshine** since there was a large chunk of missing data in it, the left missing data has been filled using mean and median respectfully depending on any possibility of invalid data noticed in analysis. We investigated the suspicous data: **Rainfall, Evaporation, WindGustSpeed, WindSpeed9am, WindSpeed3pm** that had very high difference betwen the mean and max value using *boxplot()* to look for outliers. Afte analyzing the results which shown the outliers in these data we decided to create lower and upper boundries using interquartile range for **Evaporation, WindGustSpeed, WindSpeed9am, WindSpeed3pm**, in the **Rainfall** variable we noticed deciding factor where the results above ***75*** were mostly pointing towards ***1*** in target variable **RainTomorrow**. The changes between original and outlier free data has been presented using *boxplot and histograms*. Further we decided to simplify the categorical variables related to wind directions that had a lot of unique data **WindDir9am, WindDir3pm, WindGustDir** So they show only the dominant wind direction *N, E, S and W*. We applied the changes to the date variables where we changed it into the *datetime* data type and extracted the month since it may have had the impact on rain amount and would be a usefull variable.

### Selecting the features
We created delta varibles for numerical variables to check if changes between related data have a meaning, for example: change of wind speed from 9am to 3pm has impact on rain prediction (are there any changes weather it rained or no), next we repeated the process of creating variables based on changes between related variables data for **categorical variablesfor**.

Looking at distributions of **categorical variables** we decided to use **RainToday**, **WindChange** and **WindGusDir**, because of the difference in plots between situations where **RainTommorow** is equal to ***0*** and ***1***. We decided not to include **Location** because of not that big difference in relation to number of categories in that variable. We also did not include **WindDir9am** and **WindDir3pm**, because of big possibility of correlation to already used **WindGusDir**. Initially we planned to create the varible representing the impact of particular wind direction in a certain location however this idea was dropped due to high number of variables that would be created.

Numerical variables has been chosen based on their correlation with target variable, if the correlation was above 0.09:**Sunshine, Pressure9am, Pressure3pm, Temp3pm, MaxTemp, Evaporation, WindGustSpeed, Rainfall, Cloud9am, Humidity9am, DeltaHumidity, Cloud3pm, Humidity3pm, DeltaPressure**.

Later we moved on to the **Final feature selection** in which we checked the correlation between the chosen numerical variables using the *heatmap* from its analysis we noticed that there is a strong correlation between **Pressure9am** and **Pressure3pm** as well as between **MaxTemp** and **Temp3pm**, from which we decided to drop the **Pressure9am** and **Temp3pm**. Before spliting the features we decided to check again if there is no new high correlations between the variables.

### Balancing and Standardizing
Balanced the data. Our categorical features are **Month, RainToday, WindChange, WindGustDir, RainTomorrow** using the *SMOTENC* later we One-Hot encoded the categorical data in order to be applied in the models. We decided to standardize the features using *MinMaxScaler* in order to avoid any strong changes to the original distribution, nontheless we checked again using *describe()* to be sure that there are no outstanding results and created the histogram for target value in order to check if it is balanced now. After no disturbing results has been spotted with balanced target value we proceeded to **split** the data into **train - 70%** and **test - 30%** later checking again with *describe()*.

### Training with different models
